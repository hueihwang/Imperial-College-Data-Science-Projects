```python
from google.colab import auth
auth.authenticate_user()

import gspread
from google.auth import default
creds, _ = default()
gc = gspread.authorize(creds)

from google.colab import drive
drive.mount('/content/drive')
```

```python
worksheet = gc.open("insurance").sheet1
from gspread_dataframe import get_as_dataframe

df = get_as_dataframe(worksheet)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
X = df[['age']]
y = df['charges']

plt.figure(figsize=(8, 6))
plt.scatter(X, y, alpha=0.5, color='orange')
plt.xlabel('Age')
plt.ylabel('Insurance Charges')
plt.title('Scatter Plot of Age vs. Insurance Charges')
plt.grid(True)
plt.show()
```

```python
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X, y, alpha=0.5, color='orange', label='Actual Data')
plt.plot(X, y_pred, color='red', label='Best-Fit Line')
plt.xlabel('Age')
plt.ylabel('Insurance Charges')
plt.title('Age vs. Insurance Charges with Best-Fit Line')
plt.legend()
plt.grid(True)
plt.show()
```
